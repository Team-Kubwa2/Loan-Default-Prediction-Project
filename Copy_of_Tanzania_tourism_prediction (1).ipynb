```python
import numpy as np

numeric_cols = train.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    Q1 = train[col].quantile(0.25)
    Q3 = train[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = train[(train[col] < lower_bound) | (train[col] > upper_bound)]

    print(f"{col}: {outliers.shape[0]} outliers")
```

```python
# Define a function to cap outliers using IQR
def cap_outliers(col):
    Q1 = train[col].quantile(0.25)
    Q3 = train[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    train[col] = train[col].clip(lower, upper)

# Apply to night columns
cap_outliers('total_female')
cap_outliers('total_male')
```

```python
tour_map = {
    'Independent': 0,
    'Package Tour': 1
}

train['tour_arrangement'] = train['tour_arrangement'].map(tour_map)
binary_cols = [
    'package_transport_int',
    'package_accomodation',
    'package_food',
    'package_transport_tz',
    'package_sightseeing',
    'package_guided_tour',
    'package_insurance',
    'first_trip_tz'
]

# Convert responses to lowercase to avoid case mismatch
for col in binary_cols:
    train[col] = train[col].astype(str).str.strip().str.capitalize()
    train[col] = train[col].map({'Yes': 1, 'No': 0})

print(train[binary_cols + ['tour_arrangement']].head())
```
